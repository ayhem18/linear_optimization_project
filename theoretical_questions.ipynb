# Question1
Express the following problem as a linear optimization problem 

$
\begin{align}
\min_{x \in \mathbb{R}^p} || A \cdot x - y || ^ {(1)} \\ 
0 \leq x \leq 1
\end{align}
$

A linear optimization problem is any optimization problem that can be expressed as 

$
\begin{align}
\min_{x \in \mathbb{R}^p} ~~ c^T \cdot x \\ 
W \cdot x \geq b
\end{align}
$

\begin{align*}
|| A \cdot x - y || ^ {(1)} &= \sum_{i = 1} ^ p |(A \cdot x)_i  - y_i| \text{~~using $||z||^ { ^ {(1)}} = \sum_{i = 1} ^ p |z_i| $}  \\
                                &= \sum_{i = 1} ^ m t_i \text{ where $t_i = |a_i ^ T \cdot x - y_i|$}  \\    
\end{align*}

Therefore, 

\begin{equation}
    \begin{cases}
    \min_{x \in \mathbb{R}^p} || A \cdot x - y || ^ {(1)} \\ 
    0 \leq x \leq 1
    \end{cases}
\end{equation}

is equivalent to
$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i \geq a_i^T \cdot x  - y_i ~~ i = 1, 2, ... m\\ 
        t_i \geq y_i - a_i^T \cdot x  ~~ i = 1, 2, ... m\\ 
        -x \geq -1 \\
        x \geq 0 \\
\end{cases}
\end{equation}
$

is equivalent to

$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i - a_i^T \cdot x \geq   - y_i   ~~ i = 1, 2, ... m\\ 
        t_i + a_i^T \cdot x \geq y_i    ~~ i = 1, 2, ... m\\ 
        -x \geq -1 \\
        x \geq 0 \\
\end{cases}
\end{equation}
$


Let's denote

$$
t = 
\begin{bmatrix} t_1 & t_2 &   .. & t_m 
\end{bmatrix} ^ T
$$

Then 

\begin{align*}
        t_i - a_i^T \cdot x  &\geq  - y_i ~~ i = 1, 2, ... m \\ 
        \iff \begin{bmatrix} -A & I_m\end{bmatrix} \cdot \begin{bmatrix} x \\ t\end{bmatrix} &\geq -y_i \\  
\end{align*}

similarly


\begin{align*}
        t_i + a_i^T \cdot x &\geq y_i  ~~ i = 1, 2, ... p \\ 
        \iff \begin{bmatrix} A & I_m\end{bmatrix} \cdot \begin{bmatrix} x \\ t\end{bmatrix} &\geq y_i \\  
\end{align*}


Hence The minimization problem can be expressed as: 


$$ 
\begin{cases}
    \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \begin{bmatrix} 1 & 1 & .. & 1 \end{bmatrix} ^ T \cdot t  \\ 
    W \cdot \begin{bmatrix} x \\ t\end{bmatrix} \geq b
\end{cases}
$$

where 

$$
W = 
\begin{bmatrix} 
-A & I_m \\
A & I_m \\
-I_p & 0 \\
I_p & 0 \\
\end{bmatrix}
$$

and 

$$
b = 
\begin{bmatrix} 
-y \\
y \\ 
-1_p \\
0_p  \\
\end{bmatrix}
$$

$0_p = \begin{bmatrix} 0 & 0 & ... & 0\end{bmatrix} ^ T$  with $p$ entries


# Question2 

Convert the problem to the standard format. 

A linear problem in the standard format is expressed as

$ 
\begin{equation}
    \begin{cases}
    \min_{x \in \mathbb{R}^p} c ^ T \cdot x \\ 
    W \cdot x = b \\
    x \geq 0
\end{cases}
\end{equation}
$ 

There are two necessary steps to convert a linear optimization problem into the standard format:

1. break any free variable $x_i$ into two variables $x_i^+ - x_i^-$ where $x_i^+$ and $x_i^-$ are both non-negative
2. add slack variables

Recall that our problem is equivalent to 

$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i - a_i^T \cdot x \geq   - y_i   ~~ i = 1, 2, ... m\\ 
        t_i + a_i^T \cdot x \geq y_i    ~~ i = 1, 2, ... m\\ 
        -x \geq -1 \\
        x \geq 0 \\
\end{cases}
\end{equation}
$

we know that $t_i$ are non-negative (since $t_i = |a_i^t \cdot x - y_i|$) and $x_i$ as well. So, we just need to add slack variables to the 1st, 2nd and 3th groups of constraints. 

More formally: 

$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i - a_i^T \cdot x  - s_{1i} =   - y_i   ~~ ;i = 1, 2, ... m\\ 
        t_i + a_i^T \cdot x - s_{2i} = y_i    ~~ ;i = 1, 2, ... m\\ 
        -x_i - s_{3i} = -1 ~~ ;i = 1, 2, ... p\\
        x, t, s_1, s_2, s_3\geq 0 \\
\end{cases}
\end{equation}
$

now we are ready for the matrix form: 

$$ 
\begin{cases}
    \min_{t, s_1, s_2 \in \mathbb{R}^m, x, s_3 \in \mathbb{R}^p} \begin{bmatrix} 1 & 1 & .. & 1 \end{bmatrix} ^ T \cdot t\\ 
    W \cdot \begin{bmatrix} x \\ t \\ s_1 \\ s_2 \\ s_3\end{bmatrix} = b \\
    x, t, s_1, s_2, s_3\geq 0 \\
\end{cases}
$$

$$
W = 
\begin{bmatrix} 
-A & I_m  & -I_m & 0_{m, m} & 0_{m, p}\\
A & I_m  & 0_{m, m} & -I_m & 0_{m, p}\\
-I_p & 0  & 0_{p, m} & 0_{p, m} & -I_p\\
\end{bmatrix}
$$

and 

$$
b = 
\begin{bmatrix} 
-y \\
y \\ 
-1_p
\end{bmatrix}
$$


# Question 4:

Is the solution obtained a vertex of the corresponding polyhedron ?

There are 2 possible ways to answer this question either purely algorithmically or based on the theoretical properties of the standard form and the Simplex Method.

## Algorithmic Solution: 

with $W$ of dimensions $(m, p)$, a vertex of the polyhedreon is equivalent to a Basic Feasible Solution. A BFS $x$ can be expressed as 

$[x_B, x_N] $ where $x_B \in \mathbb{R}^m $ represents the basis coordinates and $x_N \in \mathbb{R}^{n - m}$ represents the null coordinates. we have $x_N = 0$ and the column corresponding to $x_B$  in $W$ linearly independent. It is worthy to note that $x_B$ can also have zero values (in case of a degenerate vertex).

Hence, One possible algorithm to determine whether a solution is a vertex is as follows: 

1. Given a point $x$, consider all the zero entries $I$ (it is possible to have a degenerate vertex...) let's say $K$ zero entries
2. consider all combinations $ n - m \choose K$
    
    3. for each combination: consider the complement set of entries (the other unselected $m$ entries)
        
        4. consider the corresponding submatrix
        
        5. if the submatrix is invertible, then the solution is a bfs (if there is more than one possible combination, then the vertex is degenerate...)

6. reaching this point of the algorithm means that no solution is a BFS and hence there is no vertex. 


The algorithm, although general, has an exponential worst-time complexity and was not implemented since in our very specific case the obtained solution has a lot of zero entries which leads making the number $ n - m \choose K$ too large.

## Theory-based Solution:
Given a linear optimization problem in a standard form : 

$ 
\begin{equation}
    \begin{cases}
    \min_{x \in \mathbb{R}^p} c ^ T \cdot x \\ 
    W \cdot x = b \\
    x \geq 0
\end{cases}
\end{equation}
$ 

The fundamental theory of Linear Optimization states the following: 

1. if a linear problem has a non-empty feasible set, a finite optimal value, and a vertex, then there is a optimal solution that is a vertex.
2. if the linear problem is in the standard form where $W$ is full row rank, then the corresponding polyhedron does contain a vertex.
3. The simplex method moves from one BFS to another. Therefore, an optimal solution found by the Simplex method must be a vertex.

As for assumption (1), the feasible set is non-empty since the algorithm is returning at least one value. We can see that the optimal cost is finite since it is expressed as $\sum_{i = 1} ^ n t_i$ where $t_i >= 0$ making the optimal cost function at least $0$.

As for assumption (2), the problem is indeed in the standard form and 

$$
W = 
\begin{bmatrix} 
-A & I_m  & -I_m & 0_{m, m} & 0_{m, p}\\
A & I_m  & 0_{m, m} & -I_m & 0_{m, p}\\
-I_p & 0  & 0_{p, m} & 0_{p, m} & -I_p\\
\end{bmatrix}
$$

which is indeed full row-rank: the last $(2 \cdot m + p)$ columns constitute an invertible matrix $ - I_{2 \cdot m + p}$.

Hence, the obtained solution must be a vertex.
