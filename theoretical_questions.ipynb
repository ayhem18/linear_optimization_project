# Question1
Express the following problem as a linear optimization problem 

$
\begin{align}
\min_{x \in \mathbb{R}^p} || A \cdot x - y || ^ {(1)} \\ 
0 \leq x \leq 1
\end{align}
$

A linear optimization problem is any optimization problem that can be expressed as 

$
\begin{align}
\min_{x \in \mathbb{R}^p} ~~ c^T \cdot x \\ 
W \cdot x \geq b
\end{align}
$

Using $$ ||z||^ { ^ {(1)}} = \sum_{i = 1} ^ p |z_i| $$

\begin{align*}
    || A \cdot x - y || ^ {(1)} &= \sum_{i = 1} ^ p |(A \cdot x)_i  - y_i| \text{using $||z||^ { ^ {(1)}} = \sum_{i = 1} ^ p |z_i| $}  \\
                                &= \sum_{i = 1} ^ p t_i \text{ where $t_i = |a_i ^ T \cdot x - y_i|$}  \\    
\end{align*}


\begin{equation}
    \begin{cases}
    \min_{x \in \mathbb{R}^p} || A \cdot x - y || ^ {(1)} \\ 
    0 \leq x \leq 1
    \end{cases}
\end{equation}

is equivalent to
$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i \geq a_i^T \cdot x  - y_i ~~ i = 1, 2, ... m\\ 
        t_i \geq y_i - a_i^T \cdot x  ~~ i = 1, 2, ... m\\ 
        x \geq 0 \\
        -x \geq -1 \\
\end{cases}
\end{equation}
$

is equivalent to

$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^p, x \in \mathbb{R}^p} \sum_{i = 1} ^ p t_i \\
        t_i - a_i^T \cdot x \geq   - y_i   ~~ i = 1, 2, ... m\\ 
        t_i + a_i^T \cdot x \geq y_i    ~~ i = 1, 2, ... m\\ 
        x \geq 0 \\
        -x \geq -1 \\
\end{cases}
\end{equation}
$

Let's denote

$$
t = 
\begin{bmatrix} t_1 & t_2 &   .. & t_p 
\end{bmatrix} ^ T
$$

Then 

\begin{align*}
        t_i - a_i^T \cdot x  &\geq  - y_i ~~ i = 1, 2, ... p \\ 
        \iff \begin{bmatrix} -A & I_m\end{bmatrix} \cdot \begin{bmatrix} x \\ t\end{bmatrix} &\geq -y_i \\  
\end{align*}

similarly


\begin{align*}
        t_i + a_i^T \cdot x &\geq y_i  ~~ i = 1, 2, ... p \\ 
        \iff \begin{bmatrix} A & -I_m\end{bmatrix} \cdot \begin{bmatrix} x \\ t\end{bmatrix} &\geq y_i \\  
\end{align*}


Hence The minimization problem can be expressed as: 


$$ 
\begin{cases}
    \min_{t \in \mathbb{R}^p, x \in \mathbb{R}^p} \begin{bmatrix} 1 & 1 & .. & 1 \end{bmatrix} ^ T \cdot t  \\ 
    W \cdot \begin{bmatrix} x \\ t\end{bmatrix} \geq b
\end{cases}
$$

where 

$$
W = 
\begin{bmatrix} 
-A & I_m \\
A & -I_m \\
-I_p & 0 \\
I_p & 0 
\end{bmatrix}
$$

and 

$$
b = 
\begin{bmatrix} 
-y \\
y \\ 
0_p  \\
-1_p
\end{bmatrix}
$$

$0_p = \begin{bmatrix} 0 & 0 & ... & 0\end{bmatrix} ^ T$  with $p$ entries
