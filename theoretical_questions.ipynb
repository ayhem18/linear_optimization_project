# Question1
Express the following problem as a linear optimization problem 

$
\begin{align}
\min_{x \in \mathbb{R}^p} || A \cdot x - y || ^ {(1)} \\ 
0 \leq x \leq 1
\end{align}
$

A linear optimization problem is any optimization problem that can be expressed as 

$
\begin{align}
\min_{x \in \mathbb{R}^p} ~~ c^T \cdot x \\ 
W \cdot x \geq b
\end{align}
$

Using $$ ||z||^ { ^ {(1)}} = \sum_{i = 1} ^ p |z_i| $$

\begin{align*}
|| A \cdot x - y || ^ {(1)} &= \sum_{i = 1} ^ p |(A \cdot x)_i  - y_i| \text{~~using $||z||^ { ^ {(1)}} = \sum_{i = 1} ^ p |z_i| $}  \\
                                &= \sum_{i = 1} ^ m t_i \text{ where $t_i = |a_i ^ T \cdot x - y_i|$}  \\    
\end{align*}


\begin{equation}
    \begin{cases}
    \min_{x \in \mathbb{R}^p} || A \cdot x - y || ^ {(1)} \\ 
    0 \leq x \leq 1
    \end{cases}
\end{equation}

is equivalent to
$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i \geq a_i^T \cdot x  - y_i ~~ i = 1, 2, ... m\\ 
        t_i \geq y_i - a_i^T \cdot x  ~~ i = 1, 2, ... m\\ 
        -x \geq -1 \\
        x \geq 0 \\
\end{cases}
\end{equation}
$

is equivalent to

$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i - a_i^T \cdot x \geq   - y_i   ~~ i = 1, 2, ... m\\ 
        t_i + a_i^T \cdot x \geq y_i    ~~ i = 1, 2, ... m\\ 
        -x \geq -1 \\
        x \geq 0 \\
\end{cases}
\end{equation}
$


Let's denote

$$
t = 
\begin{bmatrix} t_1 & t_2 &   .. & t_m 
\end{bmatrix} ^ T
$$

Then 

\begin{align*}
        t_i - a_i^T \cdot x  &\geq  - y_i ~~ i = 1, 2, ... m \\ 
        \iff \begin{bmatrix} -A & I_m\end{bmatrix} \cdot \begin{bmatrix} x \\ t\end{bmatrix} &\geq -y_i \\  
\end{align*}

similarly


\begin{align*}
        t_i + a_i^T \cdot x &\geq y_i  ~~ i = 1, 2, ... p \\ 
        \iff \begin{bmatrix} A & I_m\end{bmatrix} \cdot \begin{bmatrix} x \\ t\end{bmatrix} &\geq y_i \\  
\end{align*}


Hence The minimization problem can be expressed as: 


$$ 
\begin{cases}
    \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \begin{bmatrix} 1 & 1 & .. & 1 \end{bmatrix} ^ T \cdot t  \\ 
    W \cdot \begin{bmatrix} x \\ t\end{bmatrix} \geq b
\end{cases}
$$

where 

$$
W = 
\begin{bmatrix} 
-A & I_m \\
A & I_m \\
-I_p & 0 \\
I_p & 0 \\
\end{bmatrix}
$$

and 

$$
b = 
\begin{bmatrix} 
-y \\
y \\ 
-1_p \\
0_p  \\
\end{bmatrix}
$$

$0_p = \begin{bmatrix} 0 & 0 & ... & 0\end{bmatrix} ^ T$  with $p$ entries


# Question2 

Convert the problem to the standard format. 

A linear problem in the standard format is expressed as

$ 
\begin{equation}
    \begin{cases}
    \min_{x \in \mathbb{R}^p} c ^ T \cdot x \\ 
    W \cdot x = b \\
    x \geq 0
\end{cases}
\end{equation}
$ 

There are two necessary steps to convert a linear optimization problem into the standard format:

1. break any free variable to $x_i$ into two variables $x_i^+ - x_i^-$ where $x_i^+$ and $x_i^-$ are both non-negative
2. add slack variables

Recall that our problem is equivalent to 

$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i - a_i^T \cdot x \geq   - y_i   ~~ i = 1, 2, ... m\\ 
        t_i + a_i^T \cdot x \geq y_i    ~~ i = 1, 2, ... m\\ 
        -x \geq -1 \\
        x \geq 0 \\
\end{cases}
\end{equation}
$

we know that $t_i$ are non-negative (since $t_i = |a_i^t \cdot x - y_i|$) and $x_i$ as well. So, we just need to add slack variables will be added to the 1st, 2nd and 3th groups of constraints. 

More formally: 

$
\begin{equation}
    \begin{cases}
        \min_{t \in \mathbb{R}^m, x \in \mathbb{R}^p} \sum_{i = 1} ^ m t_i \\
        t_i - a_i^T \cdot x  - s_{1i} =   - y_i   ~~ ;i = 1, 2, ... m\\ 
        t_i + a_i^T \cdot x - s_{2i} = y_i    ~~ ;i = 1, 2, ... m\\ 
        -x_i - s_{3i} = -1 ~~ ;i = 1, 2, ... p\\
        x, t, s_1, s_2, s_3\geq 0 \\
\end{cases}
\end{equation}
$

now we are ready for the matrix form: 

$$ 
\begin{cases}
    \min_{t, s_1, s_2 \in \mathbb{R}^m, x, s_3 \in \mathbb{R}^p} \begin{bmatrix} 1 & 1 & .. & 1 \end{bmatrix} ^ T \cdot t\\ 
    W \cdot \begin{bmatrix} x \\ t \\ s_1 \\ s_2 \\ s_3\end{bmatrix} = b \\
    x, t, s_1, s_2, s_3\geq 0 \\
\end{cases}
$$

$$
W = 
\begin{bmatrix} 
-A & I_m  & -I_m & 0_{m, m} & 0_{m, p}\\
A & I_m  & 0_{m, m} & -I_m & 0_{m, p}\\
-I_p & 0  & 0_{p, m} & 0_{p, m} & -I_p\\
\end{bmatrix}
$$

and 

$$
b = 
\begin{bmatrix} 
-y \\
y \\ 
-1_p
\end{bmatrix}
$$


# Question 4
Is the solution obtained a vertex of the corresponding polyhedron ? 

Given a linear optimization problem in a standard form : 

$ 
\begin{equation}
    \begin{cases}
    \min_{x \in \mathbb{R}^p} c ^ T \cdot x \\ 
    W \cdot x = b \\
    x \geq 0
\end{cases}
\end{equation}
$ 

The fundamental theory of Linear Optimization states that any linear optimization problem in standard form with a non-empty feasible set has at least one vertex. The important assumption here is that $W$ is full row-rank. For our problem, we have:

$$
W = 
\begin{bmatrix} 
-A & I_m  & -I_m & 0_{m, m} & 0_{m, p}\\
A & I_m  & 0_{m, m} & -I_m & 0_{m, p}\\
-I_p & 0  & 0_{p, m} & 0_{p, m} & -I_p\\
\end{bmatrix}
$$

which is indeed full row-rank: the last $(2 \cdot m + p)$ columns constitute an identity matrix $ - I_{2 \cdot m + p}$, and we have a non-empty feasible set. Hence 

Hence the optimal solution provided by the algorithm must be a vertex.
